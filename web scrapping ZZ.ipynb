```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
#scrap images from a website
var = requests.get('https://backroadsliving.com/history-of-butterflies-or-the-butterfly-gallery/').text
var
```

```python
soup = BeautifulSoup(var, 'lxml')
soup
```

```python
print(soup.prettify())
```

```python
images = soup.find_all('img')
images
```

```python
isrcs = [ ]
for i in images:
    isrcs.append(i.get('src'))
print(isrcs)
```

```python
for i in isrcs:
    webs = requests.get(i)
    open('images/' + i.split('/')[-1], 'wb').write(webs.content)
    
```
